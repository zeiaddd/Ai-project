```python
#checking outliers for transaction amount
Q1 = Ftrain['Amount'].quantile(0.25)
Q3 = Ftrain['Amount'].quantile(0.75)
IQR = Q3-Q1

lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

print(upper_bound)
print(lower_bound)
```

```python
cnt=0
for i in range(len(Ftrain)):
  if Ftrain['Amount'].iloc[i] > upper_bound:
          cnt+=1
  if Ftrain['Amount'].iloc[i] < lower_bound:
          cnt+=1
print(cnt)
```

```python
majority_class = Ftrain[Ftrain['is_fraud'] == 0]
minority_class = Ftrain[Ftrain['is_fraud'] == 1]

#Determine the number of samples needed to balance the classes
n_majority = len(majority_class)
n_minority = len(minority_class)
n_samples_needed = n_majority - n_minority

#Perform bootstrapping on the minority class
bootstrap_samples = minority_class.sample(n=n_samples_needed, replace=True, random_state=42)

#Combine the original data with the bootstrapped samples
Ftrain = pd.concat([Ftrain, bootstrap_samples])

#Shuffle the combined dataset
balanced_Ftrain =Ftrain.sample(frac=1, random_state=42).reset_index(drop=True)

#Display the balanced dataset
print(balanced_Ftrain)
```

```python
#Pie Chart
target_counts = balanced_Ftrain['is_fraud'].value_counts()


fig1 = px.pie(names=target_counts.index, values=target_counts.values,
             title='Counts of target values in train data', labels={'names': 'is_fraud'})
fig1.show()
```

```python
majority_class = Fraud_test[Fraud_test['is_fraud'] == 0]
minority_class = Fraud_test[Fraud_test['is_fraud'] == 1]

#Determine the number of samples needed to balance the classes
n_majority = len(majority_class)
n_minority = len(minority_class)
n_samples_needed = n_majority - n_minority

#Perform bootstrapping on the minority class
bootstrap_samples = minority_class.sample(n=n_samples_needed, replace=True, random_state=42)

#Combine the original data with the bootstrapped samples
Fraud_test = pd.concat([Fraud_test, bootstrap_samples])

#Shuffle the combined dataset
balanced_Fraud_test =Fraud_test.sample(frac=1, random_state=42).reset_index(drop=True)

#Display the balanced dataset
print(balanced_Fraud_test)
```

```python
#Pie Chart
target_counts = balanced_Fraud_test['is_fraud'].value_counts()


fig2 = px.pie(names=target_counts.index, values=target_counts.values,
             title='Counts of target values in test data', labels={'names': 'is_fraud'})
fig2.show()
```

```python
#separating features (X train) and target variable (Y train)
X_train = balanced_Ftrain.drop(columns=['is_fraud'])
Y_train = balanced_Ftrain['is_fraud']

X_test = balanced_Fraud_test.drop(columns=['is_fraud'])
Y_test = balanced_Fraud_test['is_fraud']


# Initialize logistic regression model
model = LogisticRegression()

# Train the model
model.fit(X_train, Y_train)

# Make predictions on test data
Y_pred = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(Y_test, Y_pred)

# Calculate F1 score
f1 = f1_score(Y_test, Y_pred)

# Calculate precision
precision = precision_score(Y_test, Y_pred)

# Calculate recall
recall = recall_score(Y_test, Y_pred)

# Print the metrics
print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Precision:", precision)
print("Recall:", recall)
```

```python
LogisticRegression_conf_matrix = confusion_matrix(Y_test, Y_pred)

print("Confusion Matrix:")
print(LogisticRegression_conf_matrix)
```

```python
#Support Vector Machine with hyperparameter tuning
svm_params = {
    'C': [0.1, 1, 10],  # Regularization.
    'gamma': [0.1, 1, 10],  #To control the similarity rad ius when training data.
    'random_state': [0] #Ensure same output with each run.
}
Svm = GridSearchCV(SVC(), svm_params, cv=5, n_jobs=-1)
Svm.fit(X_train, Y_train)

#Print the best parameters
print("Best Parameters:", Svm.best_params)

#Predict on the test set
SVM_pred = Svm.predict(X_test)

## Evaluate the model
accuracy = accuracy_score(Y_test, SVM_pred)
precision = precision_score(Y_test, SVM_pred)
recall = recall_score(Y_test, SVM_pred)
f1 = f1_score(Y_test, SVM_pred)

# Print the evaluation metrics
print("Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
SVM_conf_matrix = confusion_matrix(Y_test, SVM_pred)

print("Confusion Matrix:")
print(SVM_conf_matrix)
```

```python
# Initialize the XGBoost classifier
XGboost = XGBClassifier(n_estimators=2, max_depth=2, learning_rate=1, objective='binary:logistic')

# Train the XGBoost model
XGboost.fit(X_train, Y_train)

# Predict on the test data
preds = XGboost.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(Y_test, preds)
precision = precision_score(Y_test, preds)
recall = recall_score(Y_test, preds)
f1 = f1_score(Y_test, preds)

# Print the evaluation metrics
print("Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
XGboost_conf_matrix = confusion_matrix(Y_test, preds)

print("Confusion Matrix:")
print(XGboost_conf_matrix)
```

```python
RF_params = {
    'n_estimators': [100, 200, 300],  # Number of trees
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'random_state': [0]  # Ensure same output with each run
}

#GridSearchCV to find the best parameters
random_forest = GridSearchCV(RandomForestClassifier(), RF_params, cv=5, scoring='accuracy', n_jobs=-1)
random_forest.fit(X_train, Y_train)

#Predict on the test set
RF_predictions_test = random_forest.predict(X_test)

#Calculate and print the accuracy
accuracy = accuracy_score(Y_test, RF_predictions_test)
print("Random Forest Accuracy:", accuracy * 100,'%')

#Calculate and print the precision, recall, and F1-score
precision = precision_score(Y_test, RF_predictions_test)
recall = recall_score(Y_test, RF_predictions_test)
f1 = f1_score(Y_test, RF_predictions_test)

print("Random Forest Precision:", precision)
print("Random Forest Recall:", recall)
print("Random Forest F1-score:", f1)

best_params = random_forest.best_params
print("Best Parameters for Random Forest:", best_params)
```

```python
RF_conf_matrix = confusion_matrix(Y_test, RF_predictions_test)

print("Confusion Matrix:")
print(RF_conf_matrix)
```
